```python
# 'comments' list-> 문자열로 변형
str_list = comments
lts_result = ' '.join(s for s in str_list)
```

```python
# 3. 문단 -> 문장 (sentences)
ex_sent = kkma.sentences(lts_result);ex_sent
```

```python
# 4. 문단 -> 형태소 추출 
ex_pos = kkma.pos(lts_result) 
ex_pos 
```

```python
# 'NNG 일반 명사, NNP 고유 명사, NP 대명사' 만 추출하기 
text_data = [] 
for (text, tclass) in ex_pos : # ('형태소', 'NNG') 
    if tclass == 'NNG' or tclass == 'NNP' or tclass == 'NP' : 
        text_data.append(text)   
```

```python
text_data
```

```python
from re import match  # 전처리 위해서 정규표현식 관련 re 패키지 import
nouns = []
for sent in ex_sent : 
    for noun in kkma.nouns(sent) :
        # 단어 전처리 : 2음절 이상, 수사 제외
        if len(str(noun)) >= 2 and not(match('^[0-9]', noun)) :
            nouns.append(noun)
```

```python
word_count = {} # 빈 set

for noun in nouns :
    word_count[noun] = word_count.get(noun, 0) + 1
```

```python
from collections import Counter

counter = Counter(word_count)

top50 = counter.most_common(50);top50
```

```python
# wordcloud
import pytagcloud
import webbrowser
```

```python
# tag에 color, size, tag 사전 구성 
word_count_list = pytagcloud.make_tags(top50, maxsize=80)
# maxsize : 최대 글자크기
print(word_count_list)
```

```python
pytagcloud.create_tag_image(word_count_list,
                            'wordcloud.jpg', 
                            size=(900, 600), 
                            fontname='HANBatang', rectangular=False)
```
